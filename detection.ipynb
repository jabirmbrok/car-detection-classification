```python
import cv2

cap = cv2.VideoCapture('traffic_test.mp4')  # Ganti dengan path ke video kamu
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame) 
    result = results[0] 

    print(f"Number of car: {len(result.boxes)}")
    if len(result.boxes) > 0:
       
        for box in result.boxes.xywh.cpu().numpy(): 
            x, y, w, h = box
            x1, y1 = int(x - w / 2), int(y - h / 2)
            x2, y2 = int(x + w / 2), int(y + h / 2)
          
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)  

            label = f"{result.names[int(result.boxes.cls.cpu().numpy()[0])]} {result.boxes.conf.cpu().numpy()[0]:.2f}"
            cv2.putText(frame, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
        cv2.imshow('Car Detection', frame)
    else:
        print("No object detected.") 

    # press 'q' uto exit the video
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

Fine Tuning

```python
from ultralytics import YOLO

# Memuat model YOLOv8 dari pretrained weights
model = YOLO("yolo9c.pt")  # Ganti dengan model yang sesuai jika diperlukan

# Display model information (optional)
model.info()

# Melatih model dengan hyperparameter tuning dan augmentasi data
results = model.train(
    data="datasets/coco_car/data.yaml",  # Path ke file data.yaml
    epochs=100,                          # Jumlah epoch pelatihan
    imgsz=640,                           # Ukuran gambar input
    batch=8,                             # Ukuran batch
    lr0=0.001,                            # Learning rate awal
    momentum=0.937,                       # Momentum
    weight_decay=0.0005,                  # Weight decay
    warmup_epochs=3,                     # Warm-up epoch
    patience=10,                          # Early stopping jika loss tidak membaik
    augment=True,                         # Mengaktifkan augmentasi
    flipud=0.5,                           # Horizontal flip dengan probabilitas 50%
    fliplr=0.5,                           # Vertical flip dengan probabilitas 50%
    scale=(0.5),                     # Scaling gambar dengan faktor antara 0.5 dan 1.5
    shear=0.3,                            # Pemotongan gambar dengan factor 0.3
    perspective=0.0,                      # Pembiasan perspektif
    hsv_h=0.015,                         # Hue saturation variation
    hsv_s=0.7,                           # Saturation variation
    hsv_v=0.4,                           # Value variation
    #half=True,
)
```

```python
import cv2

cap = cv2.VideoCapture('traffic_test.mp4')

# Definisikan warna untuk setiap kategori
category_colors = {
    'car': (0, 255, 0),       # Hijau untuk mobil
    'motorcycle': (0, 0, 255), # Merah untuk motor
    'truck': (255, 0, 0)      # Biru untuk truk
}

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)  # Hasil deteksi dari model
    result = results[0]  # Ambil hasil pertama (untuk video, biasanya hanya 1 frame)

    print(f"Number of objects detected: {len(result.boxes)}")

    if len(result.boxes) > 0:
        for box, cls, conf in zip(result.boxes.xywh.cpu().numpy(), result.boxes.cls.cpu().numpy(), result.boxes.conf.cpu().numpy()):
            if conf < 0.8:  # Hanya tampilkan objek dengan confidence > 0.9
                continue

            x, y, w, h = box
            x1, y1 = int(x - w / 2), int(y - h / 2)
            x2, y2 = int(x + w / 2), int(y + h / 2)

            # Tentukan nama kategori berdasarkan class_id
            class_id = int(cls)
            class_name = result.names[class_id]

            # Tentukan warna untuk kategori
            color = category_colors.get(class_name, (255, 255, 255))  # Gunakan warna default putih jika tidak ditemukan kategori

            # Gambar bounding box dengan warna sesuai kategori
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)

            # Tambahkan label kategori dan confidence
            label = f"{class_name} {conf:.2f}"
            cv2.putText(frame, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

        cv2.imshow('Object Detection', frame)
    else:
        print("No objects detected.")

    # Tekan 'q' untuk keluar dari video
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
